**周期量和正弦型量**

我们来看看下面几个函数图像，$\sin(t),\frac{1}{2}\sin(2t),\frac{1}{4}\sin(3t)$，以及他们的和$\sin(t)+\frac{1}{2}\sin(2t)+\frac{1}{4}\sin(3t)$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,5))
t = np.linspace(-np.pi*8,np.pi*8,501)
plt.plot(t,np.sin(t),'r')
plt.plot(t,0.5*np.sin(t*2),'y-')
plt.plot(t,0.25*np.sin(t*3),'g--')
plt.plot(t,np.sin(t)+0.5*np.sin(t*2)+0.25*np.sin(t*3),'b')
plt.plot(t,np.sin(t)+0.5*np.sin(t*2)+0.25*np.sin(t*3)+0.125*np.sin(t*4)+0.06,'b')
plt.show()
```

将一个周期是T的已知函数$\phi(t)$表示为有限个，或者即使是无穷个正弦型量的和。
$\begin{eqnarray}
\phi(t) &=& A_0 + A_1 \sin(\omega t + \alpha_1) + A_2 \sin(2\omega t + \alpha_2) + A_3 \sin(3 \omega t +\alpha_3) + ... \\
&=& A_0 + \sum_{n=1}^{\infty}A_n \sin(n \omega t + \alpha_n) \\
\end{eqnarray}$其中$A_0,A_1,\alpha_1,A_2,\alpha_2,...$是常数。

已$2\pi$为周期的函数$f(x)$最终表示为：

$\begin{eqnarray}
f(x) &=& a_0 + (a_1 \cos x + b_1 \sin x) + (a_2 \cos 2x +b_2 \sin x) + (a_3 \cos 3x + b_3 \sin 3x + ... \\
&=& a_0 + \sum_{n=1}^{\infty}(a_n \cos nx +b_n \sin nx) \\
\end{eqnarray}$

$a_0=\frac{1}{2\pi} \int_{-\pi}^{\pi}f(x)dx$

$a_m=\frac{1}{\pi} \int_{-\pi}^{\pi}f(x)\cos mx d x \quad (m=1,2,...)$

$b_m=\frac{1}{\pi} \int_{-\pi}^{\pi}f(x)\sin mx dx \quad (m=1,2,...)$
